### Data Cleaning

```python
# splitting Type column into 2 new columns to seperate comma
df['Type'].str.split(',', expand= True)
```

```python
# naming new columns
df[['Boat Condition','Engine Type']]=df['Type'].str.split(',', expand= True)
```

```python
# checking new columns
df.head(500)
```

```python
# removing columns not needed
df_drops = df.drop(columns = ['Price','Type'])
```

```python
# changing column names
df_drops.rename(columns = {'Pricing' : 'Eur Price'}, inplace = True)
df_drops.rename(columns = {'Number of views last 7 days' : 'Views Last 7 Days'}, inplace = True)
```

```python
# checking column updates
df_drops.head(100)
```

```python
# checking for duplicates 
df_drops[df.duplicated()]
```

There are no duplicates

```python
# checking for null values
df_drops.isnull().sum()
```

```python
df_drops[df['Length'].isnull() == True]
```

There are a lot of empty columns in the Manufacturer, Material, and Engine Type columns. Since the most important factor in the anaylis is the number of views last 7 days column, blanks in the other columns will be ok.

```python
# checking for anything that seems out of place
df_drops.describe()
```

```python
# checking for mixed types
for col in df_drops.columns.tolist():
  weird = (df_drops[[col]].applymap(type) != df_drops[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_drops[weird]) > 0:
    print (col)
```

```python
# checking what every columns data types are
df_drops.info()
```

```python
# changing data types
df_drops['Manufacturer'] = df_drops['Manufacturer'].astype('str')
df_drops['Material'] = df_drops['Material'].astype('str')
df_drops['Location'] = df_drops['Location'].astype('str')
df_drops['Boat Condition'] = df_drops['Boat Condition'].astype('str')
df_drops['Engine Type'] = df_drops['Engine Type'].astype('str')
```

```python
# making sure their are no more mixed data types
for col in df_drops.columns.tolist():
  weird = (df_drops[[col]].applymap(type) != df_drops[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_drops[weird]) > 0:
    print (col)
```

```python
df_drops.shape
```

```python
# changing year none to nan
df_drops['Engine Type']=df_drops['Engine Type'].replace('None','nan')
```

```python
# checking changes
df_drops['Engine Type'].unique()
```

```python
#Exporting Data
df_drops.to_csv(os.path.join(path, 'Data','Prepared Data', 'boat_sales_clean.csv'))
```
